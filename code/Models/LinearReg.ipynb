```python
def MLR_dropout(property_df, dropoutNeighborhood):
    property_df_train = property_df[property_df["neighborhood"] != dropoutNeighborhood]
    del property_df_train["neighborhood"]
    property_df_test = property_df[property_df["neighborhood"] == dropoutNeighborhood]
    del property_df_test["neighborhood"]
    property_df_train_x = property_df_train.drop("total_licenses", axis=1)
    property_df_train_y = property_df_train["total_licenses"]
    
    regressor = LinearRegression()
    regressor.fit(property_df_train_x, property_df_train_y)
    
    property_df_test_x = property_df_test.drop("total_licenses", axis=1)
    property_df_test_y = property_df_test["total_licenses"]
    y_pred = regressor.predict(property_df_test_x)
    print(dropoutNeighborhood)
    print("Predicted: " + str(y_pred[0]))
    print("Actual   : " + str(property_df_test_y.values.tolist()[0]))
    print("Coeff    : " + str(regressor.coef_))
    
    return (dropoutNeighborhood, y_pred[0], property_df_test_y.values.tolist()[0])
```

```python
pred_list = []
for neighborhood in lr_dataset["neighborhood"].unique():
    pred_list.append(MLR_dropout(lr_dataset, neighborhood))
```

```python
def overEstimateFilter(pred_list_item):
    return pred_list_item[1] > pred_list_item[2]
def underEstimateFilter(pred_list_item):
    return not overEstimateFilter(pred_list_item)
```

```python
overEstimated = filter(overEstimateFilter, pred_list)
```

```python
underEstimated = filter(underEstimateFilter, pred_list)
```

```python
overEstimatedList = []
underEstimatedList = []
for x in overEstimated:
    overEstimatedList.append(x)
for x in underEstimated:
    underEstimatedList.append(x)
```
