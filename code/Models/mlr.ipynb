```python
boston_licenses = pd.read_csv(DATAFOLDER + "liquor_licenses_full.csv")
```

```python
grouped = boston_licenses.groupby(["neighborhood", "ltype"]).size()
boston_licenses["ltype"].unique()
```

```python
neighborhoods = boston_licenses["neighborhood"].unique()
data = {}
for n in neighborhoods:
    data[n] = grouped[n]
```

```python
# This just adds in 0's if a neighborhood is missing certain types of licenses
for n in data:
    for ltype in boston_licenses["ltype"].unique():
        try:
            x = data[n][ltype]
        except:
            data[n][ltype] = 0
```
