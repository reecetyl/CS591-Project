```python
# Boston Data
import pandas as pd
from geopy.geocoders import Nominatim
DATAFOLDER = "../../data/"

boston_licenses = pd.read_csv(DATAFOLDER + "boston-licenses/boston-licenses.csv")

# get rid of columns that we don't need (can always add back in later)
cols_to_delete = ["LICENSENO", "DBANAME", "STNOHI", "OPENING", "CLOSING", "PATRONSOUT", "PRIMAPPLICANT", "PHONE", "COMMENTS", "CITY", "STATE", "ZIP", "LICSTATUS", "LOCATIONCOMMENTS"]
for col in cols_to_delete:
    del boston_licenses[col]
    
# infer Location column. which locations aren't filled in?
missing_locations = boston_licenses[boston_licenses["Location"] == "(0.0, 0.0)"]

# Fill in missing locations

# Logan Airport
boston_licenses.loc[boston_licenses["Address"].str.contains("Logan"), "Location"] = "(42.366978, -71.022362)"
boston_licenses.dropna(subset=["STNO"], inplace=True)
boston_licenses
```

```python
count = 0
def address_to_lat_long(x):
    global count
    print(count)
    count += 1
    geolocator = Nominatim(user_agent="CS591 Project")
    location = geolocator.geocode(x["STNO"] + " " + x["Address"] + " Boston MA", timeout=10)
    try:
        lat, long = location.latitude, location.longitude
    except:
        return 0.0, 0.0
    return lat, long

boston_licenses = boston_licenses.astype({"STNO": int})
boston_licenses = boston_licenses.astype(str)
```

```python
# Note: this next line takes a while to interface w/ geopy API
lats = []
longs = []
for index, row in boston_licenses.iterrows():
    lat, long = address_to_lat_long(row)
    lats.append(lat)
    longs.append(long)
```

```python
boston_licenses["Lat"] = lats
boston_licenses["Long"] = longs
boston_licenses
```

```python
boston_licenses
# Take out airport
boston_licenses.loc[boston_licenses["Address"].str.contains("Logan"), "Location"] = ""
boston_licenses.drop(boston_licenses[boston_licenses.Location == ""].index, inplace=True)
# Take out CLB All Alc (Clubs)
boston_licenses.loc[boston_licenses["LICCATDESC"].str.contains("Clb. All Alc."), "Location"] = ""
boston_licenses.drop(boston_licenses[boston_licenses.Location == ""].index, inplace=True)
# Take out other clubs
boston_licenses.loc[boston_licenses["LICCATDESC"].str.contains("Club"), "Location"] = ""
boston_licenses.drop(boston_licenses[boston_licenses.Location == ""].index, inplace=True)
# Take out universities
boston_licenses.loc[boston_licenses["BUSINESSNAME"].str.contains("UNIVERSITY"), "Location"] = ""
boston_licenses.drop(boston_licenses[boston_licenses.Location == ""].index, inplace=True)

# Drop where we can't find the lat/long
boston_licenses = boston_licenses[boston_licenses["Location"] != "(0.0, 0.0)"]
boston_licenses = boston_licenses[boston_licenses["Lat"] != 0.0]

# Reduce LICCAT Types (combine types)
boston_licenses.loc[boston_licenses["LICCAT"].str.contains("GOP"), "LICCAT"] = "GOPAL"
boston_licenses.loc[boston_licenses["LICCAT"].str.contains("CLB"), "LICCAT"] = "CLB"
boston_licenses.drop(boston_licenses[boston_licenses.LICCAT == "CLB"].index, inplace=True)
boston_licenses.loc[boston_licenses["LICCAT"].str.contains("CV7"), "LICCAT"] = "CV7"
boston_licenses.loc[boston_licenses["LICCAT"].str.contains("INN"), "LICCAT"] = "INN"
boston_licenses.loc[boston_licenses["LICCAT"].str.contains("TAVAL"), "LICCAT"] = "TAV"
boston_licenses.LICCAT.unique()
boston_licenses
```

```python
# save as CSV
boston_licenses.to_csv(DATAFOLDER + "boston-licenses/boston-licenses-clean.csv", index = False)
```
